```python
import requests
from openai import OpenAI
#pip install openai
```

```python
response = requests.get("http://localhost:11434/api/tags") # default port für ollama
models = response.json()
print("Verfügbare Modelle:")
for m in models["models"]:
    print(f"  • {m['name']}")
```

```python
# Zeigt das Kernprinzip: gleicher Code wie OpenAI, andere URL
client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"  # wird ignoriert, muss aber gesetzt sein
)

response = client.chat.completions.create(
    model="gemma3:4b",  # dein lokales Modell
    messages=[{"role": "user", "content": "Sag 'Hallo LAIKA' und nichts sonst."}]
)

print(response.choices[0].message.content)
```

```python
stream = client.chat.completions.create(
    model="gemma3:4b",
    messages=[{"role": "user", "content": "gib mir 5 zufällige wörter, ein wort pro Zeile bitte"}],
    stream=True
)

for chunk in stream:
    if chunk.choices[0].delta.content:
        print(chunk.choices[0].delta.content, end="", flush=True)
```
