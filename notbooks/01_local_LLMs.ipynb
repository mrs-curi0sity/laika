```python
import requests
from openai import OpenAI
#pip install openai
```

```python
response = requests.get("http://localhost:11434/api/tags") # default port für ollama
models = response.json()
print("Verfügbare Modelle:")
for m in models["models"]:
    print(f"  • {m['name']}")
```

```python


# Zeigt das Kernprinzip: gleicher Code wie OpenAI, andere URL
client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"  # wird ignoriert, muss aber gesetzt sein
)

response = client.chat.completions.create(
    model="llama3.2",  # dein lokales Modell
    messages=[{"role": "user", "content": "Sag 'Hallo LAIKA' und nichts sonst."}]
)

print(response.choices[0].message.content)
```
