```python
# Golden Dataset für deinen Ermittlungsgraph
test_cases = [
    {"frage": "Wer ist der Anführer?",           "erwartet": "Darth Vader"},
    {"frage": "Wer hat die meisten Nachrichten?", "erwartet": "The Dude"},
    {"frage": "Welche Gruppen gibt es?",           "erwartet": "Freie Männer"},
    {"frage": "Wer kennt den V-Mann?",             "erwartet": "Jules Winnfield"},
]

from notebooks.05_agents import run_agent  # oder copy-paste

richtig = 0
for test in test_cases:
    antwort = run_agent(test["frage"])
    ok = test["erwartet"].lower() in antwort.lower()
    richtig += ok
    print(f"{'✅' if ok else '❌'} {test['frage'][:40]}")

print(f"\nScore: {richtig}/{len(test_cases)} = {richtig/len(test_cases)*100:.0f}%")
```
